# Python 基本輸出應用

**1. python預設輸出換行**

```python
print('aa')
print('bb')
print('cc')
```

**2. python若不想換行輸出，則在print後面加入end關鍵字**

```python
print('aa', end='')
print('bb')
```

 **3. python不換行輸出且中間用符號區隔**

```python
print('aa', end = ',')
print('bb', end = '/')
print('cc')
```

**4. 宣告值輸出應用 換行**

```python
a, b, c = 1, 2, 3
print(a)
print(b)
print(c)
```

**5. 宣告值輸出應用 不換行**

```python
a, b, c = 1, 2, 3
print(a, end = ',')
print(b, end = '/')
print(c)
```

**--------------------------
$ 為輸出文字控制字元
--------------------------**

**6. 指定宣告值輸出應用 整數**

```python
#整數用d
a, b, c = 1, 2, 3
print("a = %d, b = %d, c = %d" % (a, b, c))
```

**7. 指定宣告值輸出應用 浮點數**

```python
#浮點數用f
a = 3.4
print("浮點數a = %f" % a)
print("整數a = %d" % a)
```

**8. 指定宣告值輸出應用 字元**

```python
#對應ASCII Code 75 為 K

#字元用c
a = 75
print("整數a = %d" % a)
print("字元a = %c" % a)
```

**9. 指定宣告值輸出應用 字串**

```python
a = "你好"
b = 6

print("問候別人要說%s, 明天%d點見面" % (a, b))
```

**10. 指定宣告值輸出應用 宣告固定輸出範圍 整數**

```python
a = 100
b = 1000
print("%2d, %2d" % (a, b))

a = 100
b = 1000
print("%8d, %8d" % (a, b))

a = 100
b = 1000
print("%20d, %20d" % (a, b))
```

**11. 指定宣告值輸出應用 宣告固定輸出範圍 小數點**

```python
a = 3.14159
print("%2.2f" % a)

a = 3.14159
print("%10.2f" % a)

a = 3.14159
print("%2.4f" % a)

a = 3.14159
print("%10.4f" % a)
```

**12. 指定宣告值輸出應用 宣告固定輸出範圍 小數點**

```python
a = 5
b = 628
print("%+6d%+6d" % (a, b))

a = -5
b = -628
print("%-6d%-6d" % (a, b))
```

**13. 數學式應用**

```python
a = 1
b = -2
c = 3
print('%dx%+dy%+d=0' % (a, b, c))
```

# 進階COPY套件輸出

```python
import copy
```

```python
b = [1, 2, 'a', 'b', [1, 2, 3], 4]
#assign
a = b
print("物件")
print(id(b))
print(id(a))
print("子物件")
print(id(b[4]))
print(id(a[4]))
b.append(5)
print("改變物件後：")
print(b)
print(a)
b[4].append(4)
print("改變子物件後：")
print(b)
print(a)
```

```python
b = [1, 2, 'a', 'b', [1, 2, 3], 4]
#shallow copy - 1
a = b[:]
print("物件")
print(id(b))
print(id(a))
print("子物件")
print(id(b[4]))
print(id(a[4]))
b.append(5)
print("改變物件後：")
print(b)
print(a)
b[4].append(4)
print("改變子物件後：")
print(b)
print(a)
```

```python
b = [1, 2, 'a', 'b', [1, 2, 3], 4]
#shallow copy - 2
a = b.copy()
print("物件")
print(id(b))
print(id(a))
print("子物件")
print(id(b[4]))
print(id(a[4]))
b.append(5)
print("改變物件後：")
print(b)
print(a)
b[4].append(4)
print("改變子物件後：")
print(b)
print(a)
```

```python
b = [1, 2, 'a', 'b', [1, 2, 3], 4]
#deep copy
a = copy.deepcopy(b)
print("物件")
print(id(b))
print(id(a))
print("子物件")
print(id(b[4]))
print(id(a[4]))
b.append(5)
print("改變物件後：")
print(b)
print(a)
b[4].append(4)
print("改變子物件後：")
print(b)
print(a)
```

# 進階資料型態輸出

```python
a = True
b = 123
c = 1.23
d = '123'
e = 3 + 5j
f = [1, 2, 3]
g = (1, 2, 3)
h = {1, 2, 3}
i = {"蘋果" : "apple", "香蕉" : "banana"}
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
print(g)
print(type(g))
print(h)
print(type(h))
print(i)
print(type(i))
```

# 進階進制轉換輸出

```python
x = 0b1101          # 這是2進為整數
print(x)            # 列出10進位的結果
y = 13              # 這是10進為整數
print(bin(y))       # 列出轉換成2進位的結果
```

```python
x = 0o57            # 這是8進為整數
print(x)            # 列出10進位的結果
y = 47              # 這是10進為整數
print(oct(y))       # 列出轉換成8進位的結果
```

```python
x = 0x5D            # 這是16進為整數
print(x)            # 列出10進位的結果
y = 93              # 這是10進為整數
print(hex(y))       # 列出轉換成16進位的結果
```
